```python
import pandas as pd
from sklearn.metrics import accuracy_score,f1_score
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
data=pd.read_excel('Final Labeled Dataset.xlsx')
data.head()
```

## length

```python
data['Headlines_length'] = data['Headlines'].apply(lambda x: len(x))
data.head()
```

## no. of nouns

```python
import pandas as pd
import nltk
from nltk import word_tokenize, pos_tag
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('stopwords')

def count_nouns_in_headline(headline):
    # Tokenize the headline into words
    words = word_tokenize(headline)

    # Remove stop words and non-alphabetic characters
    stop_words = set(stopwords.words('english'))
    words = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]

    # Perform POS tagging to identify nouns
    tagged_words = pos_tag(words)

    # Count the number of nouns
    noun_count = sum(1 for word, tag in tagged_words if tag.startswith('NN'))

    return noun_count

# Calculate the number of nouns for each headline and store it in a new column 'Noun_Count'
data['Noun_Count'] = data['Headlines'].apply(count_nouns_in_headline)
data.head()
```

## verb count

```python
import pandas as pd
import nltk
from nltk import word_tokenize, pos_tag
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('stopwords')

def count_nouns_in_headline(headline):
    # Tokenize the headline into words
    words = word_tokenize(headline)

    # Remove stop words and non-alphabetic characters
    stop_words = set(stopwords.words('english'))
    words = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]

    # Perform POS tagging to identify nouns
    tagged_words = pos_tag(words)

    # Count the number of nouns
    noun_count = sum(1 for word, tag in tagged_words if tag.startswith('VB'))

    return noun_count

# Calculate the number of nouns for each headline and store it in a new column 'Noun_Count'
data['verb_count'] = data['Headlines'].apply(count_nouns_in_headline)
data.head()
```

## combine features

```python
text_data = data['Headlines']

# TF-IDF Vectorization of the text data
tfidf_vectorizer = TfidfVectorizer()
tfidf_features = tfidf_vectorizer.fit_transform(text_data)

# Additional numerical features (assuming 'numerical_features' contains numerical data)
numerical_features = data[['Headlines_length', 'Noun_Count', 'verb_count']]

# Convert the TF-IDF features to a dense NumPy array
tfidf_features_array = tfidf_features.toarray()

# Combine the TF-IDF features with numerical features
combined_features = np.hstack((tfidf_features_array, numerical_features))

# Now you have a NumPy array 'combined_features' with both TF-IDF and numerical features.
```

## train test

```python
x = combined_features
y = data.Claim
X_train,X_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=9000, stratify=y)
```

## AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
# Create adaboost classifer object
abc = AdaBoostClassifier(n_estimators=50,
                         learning_rate=1)
# Train Adaboost Classifer
model = abc.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```

## Decision Tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(
          max_depth=1,
          random_state=42)
tree = tree.fit(X_train, y_train)
y_test_pred = tree.predict(X_test)
```

```python
accuracy_score(y_test_pred, y_test)
```

## LR

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
logisticRegr.fit(X_train, y_train)
```

```python
predictions = logisticRegr.predict(X_test)
predictions
```

```python
accuracy_score(predictions, y_test)
```

## SVM

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```
