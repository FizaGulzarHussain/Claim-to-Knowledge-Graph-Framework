## Read OIE file

```python
import re

# Open the file in read mode
with open("input.txt", "r") as file:
    # Read the lines from the file
    lines = file.readlines()

# Initialize lists for arg1, arg2, and rel
arg1_list = []
arg2_list = []
rel_list = []

# Iterate over the lines
for line in lines:
    # Split the line into sentence and triple components
    parts = line.split("\t")
    triple = parts[1].strip()

    # Use regular expressions to extract arg1, arg2, and rel
    arg1_match = re.search(r'<arg1>(.*?)</arg1>', triple)
    arg2_match = re.search(r'<arg2>(.*?)</arg2>', triple)
    rel_match = re.search(r'<rel>(.*?)</rel>', triple)

    # Append the extracted values to the respective lists
    arg1_list.append(arg1_match.group(1))
    arg2_list.append(arg2_match.group(1))
    rel_list.append(rel_match.group(1))

# Print the extracted arg1, arg2, and rel values
for i in range(len(arg1_list)):
    print("arg1:", arg1_list[i])
    print("arg2:", arg2_list[i])
    print("rel:", rel_list[i])
    print()
```

# Read OIE6 triples 

### extract subject,object,relation and sentence from and save it

```python
import re

# Open the input file in read mode
with open("input.txt", "r") as file:
    # Read the lines from the file
    lines = file.readlines()

# Initialize lists for sentences, arg1, arg2, and rel
sentences = []
arg1_list = []
arg2_list = []
rel_list = []

# Iterate over the lines
for line in lines:
    # Split the line into sentence and triple components
    parts = line.split("\t")
    sentence = parts[0].strip()
    triple = parts[1].strip()

    # Use regular expressions to extract arg1, arg2, and rel
    arg1_match = re.search(r'<arg1>(.*?)</arg1>', triple)
    arg2_match = re.search(r'<arg2>(.*?)</arg2>', triple)
    rel_match = re.search(r'<rel>(.*?)</rel>', triple)

    # Append the extracted values to the respective lists
    sentences.append(sentence)
    arg1_list.append(arg1_match.group(1))
    arg2_list.append(arg2_match.group(1))
    rel_list.append(rel_match.group(1))

# Open the output file in write mode
with open("output.txt", "w") as file:
    # Write the extracted information to the file
    for i in range(len(sentences)):
        file.write("Sentence: {}\n".format(sentences[i]))
        file.write("arg1: {}\n".format(arg1_list[i]))
        file.write("arg2: {}\n".format(arg2_list[i]))
        file.write("rel: {}\n".format(rel_list[i]))
        file.write("\n")
```

# extract noun phrase from args

```python
import nltk

def extract_noun_phrases(sentence):
    # Tokenize the sentence into words
    words = nltk.word_tokenize(sentence)

    # Perform part-of-speech tagging
    tagged_words = nltk.pos_tag(words)

    # Define the grammar for noun phrase chunking
    grammar = r"NP: {<DT>?<JJ>*<NN.*>+}"

    # Create a chunk parser using the defined grammar
    chunk_parser = nltk.RegexpParser(grammar)

    # Parse the tagged words to extract noun phrases
    parsed_tree = chunk_parser.parse(tagged_words)

    # Extract noun phrases from the parsed tree
    noun_phrases = []
    for subtree in parsed_tree.subtrees():
        if subtree.label() == 'NP':
            noun_phrase = " ".join(word for word, tag in subtree.leaves())
            noun_phrases.append(noun_phrase)

    return noun_phrases

# Open the file for reading and writing
with open('output.txt', 'r+') as file:
    lines = file.readlines()
    file.seek(0)  # Move the file pointer to the beginning

    # Iterate over the lines and extract noun phrases
    for line in lines:
        line = line.strip()
        if line.startswith("Sentence:"):
            sentence = line.replace("Sentence: ", "")
            file.write(line + "\n")
        elif line.startswith("arg1:"):
            arg1 = line.replace("arg1: ", "")
            noun_phrases = extract_noun_phrases(arg1)
            file.write("arg1: " + "  ".join(noun_phrases) + "\n")
        elif line.startswith("arg2:"):
            arg2 = line.replace("arg2: ", "")
            noun_phrases = extract_noun_phrases(arg2)
            file.write("arg2: " + "  ".join(noun_phrases) + "\n")
        elif line.startswith("rel:"):
            file.write(line + "\n")
        else:
            file.write(line + "\n")

    file.truncate()  # Remove any remaining content from the previous file version
```

# strict match

```python
import re

# Read the DBpedia file
with open('falconDataset.txt', 'r') as dbpedia_file:
    dbpedia_content = dbpedia_file.readlines()

# Read the triples file
with open('output.txt', 'r') as triples_file:
    triples_content = triples_file.readlines()

# Create a dictionary to store the DBpedia entities and their URIs
dbpedia_entities = {}

# Parse the DBpedia content
for line in dbpedia_content:
    if line.startswith('{"entities":'):
        entities = re.findall(r'"URI":"(.*?)","surface form":"(.*?)"', line)
        for entity_uri, surface_form in entities:
            dbpedia_entities[surface_form] = entity_uri

# Process the triples content
processed_triples = []

for line in triples_content:
    if line.startswith('Sentence:'):
        sentence = line.split('Sentence:')[1].strip()
        processed_triples.append(f"Sentence: {sentence}")
    elif line.startswith('arg1:'):
        arg1 = line.split('arg1:')[1].strip()
        processed_triples.append(f"arg1: {arg1} (DBpedia URI: {dbpedia_entities.get(arg1, 'Not Found')})")
    elif line.startswith('arg2:'):
        arg2 = line.split('arg2:')[1].strip()
        processed_triples.append(f"arg2: {arg2} (DBpedia URI: {dbpedia_entities.get(arg2, 'Not Found')})")
    elif line.startswith('rel:'):
        rel = line.split('rel:')[1].strip()
        processed_triples.append(f"rel: {rel}")

# Save the processed triples to the same file
with open('ff.txt', 'w') as output_file:
    output_file.write('\n'.join(processed_triples))
```

```python
import re

# Read the DBpedia file
with open('falconDataset.txt', 'r') as dbpedia_file:
    dbpedia_content = dbpedia_file.readlines()

# Read the triples file
with open('output.txt', 'r') as triples_file:
    triples_content = triples_file.readlines()

# Create a dictionary to store the DBpedia entities and their URIs
dbpedia_entities = {}

# Parse the DBpedia content
for line in dbpedia_content:
    if line.startswith('{"entities":'):
        entities = re.findall(r'"URI":"(.*?)","surface form":"(.*?)"', line)
        for entity_uri, surface_form in entities:
            dbpedia_entities[surface_form.lower()] = entity_uri

# Process the triples content
processed_triples = []

for line in triples_content:
    if line.startswith('Sentence:'):
        sentence = line.split('Sentence:')[1].strip()
        processed_triples.append(f"Sentence: {sentence}")
    elif line.startswith('arg1:'):
        arg1 = line.split('arg1:')[1].strip()
        dbpedia_uri = next((uri for surface, uri in dbpedia_entities.items() if surface in arg1.lower()), 'Not Found')
        processed_triples.append(f"arg1: {arg1} (DBpedia URI: {dbpedia_uri})")
    elif line.startswith('arg2:'):
        arg2 = line.split('arg2:')[1].strip()
        dbpedia_uri = next((uri for surface, uri in dbpedia_entities.items() if surface in arg2.lower()), 'Not Found')
        processed_triples.append(f"arg2: {arg2} (DBpedia URI: {dbpedia_uri})")
    elif line.startswith('rel:'):
        rel = line.split('rel:')[1].strip()
        processed_triples.append(f"rel: {rel}")

# Save the processed triples to the same file
with open('triples_file.txt', 'w') as output_file:
    output_file.write('\n'.join(processed_triples))
```
