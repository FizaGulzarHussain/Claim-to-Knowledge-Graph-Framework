# Matching Function

```python
class Ref:
    def __init__(self, pred, args):
        self.pred = pred
        self.args = args


class Ex:
    def __init__(self, pred, args):
        self.pred = pred
        self.args = args


def tuple_match(ref, ex):
    precision = [0, 0]  # 0 out of 0 predicted words match
    recall = [0, 0]  # 0 out of 0 reference words match

    # If, for each part, any word is the same as a reference word, then it's a match.
    predicted_words = ex.pred.split()
    gold_words = ref.pred.split()
    precision[1] += len(predicted_words)
    recall[1] += len(gold_words)

    matching_words = 0
    for w in gold_words:
        if w in predicted_words:
            matching_words += 1
            predicted_words.remove(w)

    if matching_words == 0:
        return False  # t <-> gt is not a match
    precision[0] += matching_words
    recall[0] += matching_words

    for i in range(len(ref.args)):
        gold_words = ref.args[i].split()
        recall[1] += len(gold_words)
        if len(ex.args) <= i:
            if i < 2:
                return False
            else:
                continue
        predicted_words = ex.args[i].split()
        precision[1] += len(predicted_words)
        matching_words = 0
        for w in gold_words:
            if w in predicted_words:
                matching_words += 1
                predicted_words.remove(w)

        if matching_words == 0 and i < 2:
            return False  # t <-> gt is not a match
        precision[0] += matching_words
        # Currently this slightly penalizes systems when the reference
        # reformulates the sentence words because the reformulation doesn't
        # match the predicted word. It's a one-wrong-word penalty to precision,
        # to all systems that correctly extracted the reformulated word.
        recall[0] += matching_words

    prec = 1.0 * precision[0] / precision[1]
    rec = 1.0 * recall[0] / recall[1]
    return [prec, rec]
```

# Dependency Parsing

```python
import pandas as pd
data=pd.read_excel('DP File.xlsx')
data.head()
```

```python
precision_recall= []
for i in range(len(data)):
  ref1 = Ref(str(data['Relations '][i]), [str(data['Subject'][i]), str(data['Object'][i])])
  ex1 = Ex(str(data['predicted_relations'][i]), [str(data['Predicted_Subject'][i]), str(data['Predicted_object'][i])])
  precision_recall.append(tuple_match(ref1, ex1))

print(len(precision_recall))
```

```python
precision_recall
```

```python
filtered_list = [item for item in precision_recall if item != False]

# Print the filtered list
print(filtered_list)
```

### DP res

```python
sumP = 0
recall=0
for i in range(len(filtered_list)):
  sumP += filtered_list[i][0]
  recall += filtered_list[i][1]
sumP= sumP/5001
recall= recall/5001
```

```python
recall
```

```python
#f1
f1 = 2 * (sumP * recall) / (sumP + recall)
f1
```

```python
a1[0]
```

```python
def lexical(ref, ex):
  precision = [0, 0] # 0 out of 0 predicted words match
  recall = [0, 0] # 0 out of 0 reference words match
        # If, for each part, any word is the same as a reference word, then it's a match.

  predicted_words = ex.split()
  gold_words = ref.split()
  precision[1] += len(predicted_words)
  recall[1] += len(gold_words)
  matching_words = 0
  for w in gold_words:
    if w in predicted_words:
      matching_words += 1
      predicted_words.remove(w)

  if matching_words == 0:
    return False # t <-> gt is not a match
  precision[0] += matching_words
  recall[0] += matching_words
  prec = 1.0 * precision[0] / precision[1]
  rec = 1.0 * recall[0] / recall[1]
  return [prec, rec]
```
